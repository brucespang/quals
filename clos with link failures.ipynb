```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import fnss
```

```python
G = fnss.fat_tree_topology(6)
```

```python
nx.draw(G, with_labels=True)
```

```python
for edge in G.edges:
    u,v = edge
    if G.nodes[u]['type'] != 'host' and G.nodes[v]['type'] != 'host':
        G.edges[edge]['capacity'] = 1
```

```python
G.remove_edge(13,9)
```

```python
hosts = [id for id,node in G.nodes.items() if node['layer'] == 'leaf']
```

```python
for a1 in hosts:
    for a2 in hosts:
        if a1 == a2: continue
        shortest_paths = len(list(nx.all_shortest_paths(G, a1, a2)))
        if shortest_paths > 1:
            print(a1, a2, nx.maximum_flow_value(G, a1, a2), shortest_paths)
```

```python
paths = list(nx.all_shortest_paths(G, 49, 51))
subgraph = nx.Graph()
for path in paths:
    nx.add_path(subgraph, path)
nx.draw(subgraph, with_labels=True, font_color='white')
```

```python
paths = list(nx.all_shortest_paths(G, 49, 70))
subgraph = nx.Graph()
for path in paths:
    nx.add_path(subgraph, path)
nx.draw(subgraph, with_labels=True, font_color='white')    
```
