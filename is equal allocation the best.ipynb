```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plorts
plt.style.use(['plorts', 'plorts-web'])
```

```python
def delay(x, C=10):
    return np.exp(x-C)
```

```python
xs = np.arange(0, 15, 0.05)
plt.plot(xs, [delay(x) for x in xs])
plt.xlabel("Offered load (Mbps)")
plt.ylabel("Delay (ms)")
plorts.style_axis()
```
