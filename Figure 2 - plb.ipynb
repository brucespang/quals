```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plorts
plt.style.use(['plorts', 'plorts-web'])
```

```python
# df from https://apps.automeris.io/wpd/
df = pd.read_csv('./plb_figure_2.csv', header=None)
df = df.ffill(axis=1)
df
```

```python
df.columns = [f"{axis}{dataset}" for axis, dataset in zip(df.iloc[1], df.iloc[0])]
df = df.drop([0,1])
df
```

```python
df = pd.wide_to_long(df, ['X', 'Y'], i="id", j="dataset", suffix='Dataset \d+').reset_index()
```

```python
df = df.dropna()
```

```python
def interp_nans(x, y):
    is_nan = np.isnan(y)
    res = y * 1.0
    res[is_nan] = np.interp(x[is_nan], x[-is_nan], y[-is_nan])
    return res

def interp(df, x, y):
    df = df.sort_values(x)
    df[y] = interp_nans(df[x], df[y])
    return df[[x,y]]

def stackplot(data, x, y, hue, cmap=plorts.palettes.neon):
    combined_df = []
    for x_val in df[x]:
        for h_val in set(df[hue]):
            matches = df[(df[x] == x_val) & (df[hue] == h_val)]
            if len(matches) == 0:
                combined_df.append(pd.DataFrame([{
                    x: x_val,
                    y: None,
                    hue: h_val
                }]))
            elif len(matches) == 1:
                combined_df.append(matches)
            else:
                print(matches)
    combined_df = pd.concat(combined_df)
    
    interpolated = combined_df.groupby(hue).apply(lambda df: interp(df, x, y)).reset_index()

    xs = np.array(interpolated[x])
    yss = []
    labels = []
    for k,grp in interpolated.groupby(hue):
        labels.append(k)
        grp_xs = grp[x].tolist()
        grp_ys = grp[y].tolist()
        ys = []
        for v in xs:
            if len(grp_xs) > 0 and grp_xs[0] == v:
                ys.append(grp_ys.pop(0))
                grp_xs.pop(0)
            else:
                if len(ys) == 0:
                    ys.append(0.)
                else:
                    ys.append(ys[-1])
        assert len(grp_xs) == 0
        assert len(grp_ys) == 0
        assert len(ys) == len(xs)
        yss.append(np.array(ys, dtype=float))

    if cmap is not None:
        colors = plorts.colors_from_hue(interpolated, hue, cmap)
    else:
        colors = None

    plt.stackplot(xs, *yss, labels=labels, colors=colors)

    plt.ylabel(y)
    plt.xlabel(x)
    plt.gca().autoscale(tight=True)
    plt.gca().margins(y=0.1)
```

```python
stackplot(df, x='X', y='Y', hue='dataset')
plt.xlabel("Time")
plt.ylabel("Aggregate link utilization")
plt.axis(xmin=0, xmax=12)
plt.xticks([])
plt.axvline(x=6.5, linestyle=":", color="grey")
plt.text(2, 550, "Flowlet ECMP", fontsize=16)
plt.text(8, 550, "PLB", fontsize=16)
plorts.style_axis()
plt.savefig("fig_2.pdf")
```
